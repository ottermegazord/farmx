```python
ds_weather = pd.read_hdf("/home/ottermegazord/PycharmProjects/funstuff/farmx/data/2017_DS.h5")
sg_weather = pd.read_hdf("/home/ottermegazord/PycharmProjects/funstuff/farmx/data/2017_sembawang.h5")
```

```python
ds_weather.keys()
```

```python
sg_weather = sg_weather.drop(columns='index')
sg_weather.keys()
```

```python
def scale_plot_size(factor=1.5):
    import matplotlib as mpl
    default_dpi = mpl.rcParamsDefault['figure.dpi']
    mpl.rcParams['figure.dpi'] = default_dpi*factor

scale_plot_size(3.0)
ax = plt.axes()
plt.plot(ds_weather.index, ds_weather['temperature'], linewidth=1.0)
plt.plot(sg_weather.index, sg_weather['temperature'], linewidth=0.2)
ax.xaxis.grid(True)
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %y'))
plt.setp(ax.get_xticklabels(), rotation=85, fontsize=8)
plt.xlabel('Datetime', fontsize=8)
plt.ylabel('Rainfall (mm)', fontsize=8)
plt.title('Temperature against time (Darksky)', fontsize=8)
```
